**ALGORITMO K-MEANS**

```python
#K-Means
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Crear un conjunto de datos simulado
X, y = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=0.60)

# Visualizar el conjunto de datos
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title('Conjunto de Datos Simulado')
plt.show()

# Aplicar el algoritmo K-Means
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

# Obtener las etiquetas de clúster y los centroides
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualizar los resultados del clúster
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroides')
plt.title('Resultados del Clúster con K-Means')
plt.legend()
plt.show()
```

**Hierarchical Clustering**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import linkage, dendrogram

# Crear un conjunto de datos simulado
X, y = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=0.60)

# Aplicar el algoritmo de Hierarchical Clustering
linkage_matrix = linkage(X, method='ward')

# Visualizar el dendrograma
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix)
plt.title('Dendrograma de Hierarchical Clustering')
plt.xlabel('Índices de los Datos')
plt.ylabel('Distancia Euclidiana')
plt.show()
```

**ALGORITMO DBSCAN**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Crear un conjunto de datos simulado
X, _ = make_moons(n_samples=200, noise=0.05, random_state=42)

# Aplicar el algoritmo DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Visualizar los resultados
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis', edgecolors='k')
plt.title('Resultados de DBSCAN')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.show()
```
