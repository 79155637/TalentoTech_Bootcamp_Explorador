**GRAFICOS BASICOS**

```python
#Histogramas

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Crear un DataFrame aleatorio
np.random.seed(42)
data = {'columna': np.random.randn(1000)}
df = pd.DataFrame(data)

# Ejemplo: Histograma de una variable
sns.histplot(data=df['columna'], kde=False, bins=20, color='skyblue')
plt.title('Histograma de la Variable')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()


#KDE

import seaborn as sns
import matplotlib.pyplot as plt

# Ejemplo: Histograma y KDE de una variable
sns.histplot(data=df['columna'], kde=True, color='skyblue')
plt.title('Histograma y KDE de la Variable')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()


#SCATTER PLOTS

import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame aleatorio con dos variables
np.random.seed(42)
data = {'columna_x': np.random.randn(100), 'columna_y': 2 * np.random.randn(100) + 0.5}
df = pd.DataFrame(data)

# Ejemplo: Scatter plot de dos variables
sns.scatterplot(x='columna_x', y='columna_y', data=df, color='blue')
plt.title('Scatter Plot de Dos Variables')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()


#REGRESION LINEAL

import seaborn as sns
import matplotlib.pyplot as plt

# Ejemplo: Scatter plot con regresión lineal
sns.regplot(x='columna_x', y='columna_y', data=df, color='blue')
plt.title('Scatter Plot con Regresión Lineal')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()


```

**PERSONALIZACION Y ESTILO**

```python
# PALETA DEEP

import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame aleatorio con dos variables
np.random.seed(42)
data = {'columna_x': np.random.randn(100), 'columna_y': 2 * np.random.randn(100) + 0.5}
df = pd.DataFrame(data)

# Utilizar una paleta de colores predefinida
sns.set_palette("deep")

# Ejemplo: Scatter plot con la paleta "deep"
sns.scatterplot(x='columna_x', y='columna_y', data=df)
plt.title('Scatter Plot con Paleta "deep"')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()

#PERSONALIZACION PALETAS

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame aleatorio con dos columnas
np.random.seed(42)
data = {'categorias': np.random.choice(['A', 'B', 'C'], size=100),
        'valores': np.random.randn(100)}
df = pd.DataFrame(data)

# Crear una paleta personalizada
mi_paleta = sns.color_palette("husl", 8)

# Utilizar la paleta personalizada
sns.set_palette(mi_paleta)

# Ejemplo: Bar plot con la paleta personalizada
sns.barplot(x='categorias', y='valores', data=df)
plt.title('Bar Plot con Paleta Personalizada')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()


#AJUSTES DE ESTILOS

import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame aleatorio
np.random.seed(42)
data = {'columna': np.random.randn(1000)}
df = pd.DataFrame(data)

# Utilizar un estilo predefinido
sns.set_style("darkgrid")

# Ejemplo: Histograma con el estilo "darkgrid"
sns.histplot(data=df['columna'], kde=False, bins=20, color='skyblue')
plt.title('Histograma con Estilo "darkgrid"')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()


#PERSONZALICION DE ESTILOS

import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame aleatorio con dos columnas
np.random.seed(42)
data = {'categoria': np.random.choice(['A', 'B', 'C'], size=100),
        'valor': np.random.randn(100)}
df = pd.DataFrame(data)

# Personalizar el estilo
sns.set(style="whitegrid", rc={"grid.linestyle": "--", "axes.labelsize": 14, "xtick.labelsize": 12, "ytick.labelsize": 12})

# Ejemplo: Box plot con estilo personalizado
sns.boxplot(x='categoria', y='valor', data=df, palette='pastel')
plt.title('Box Plot con Estilo Personalizado')
plt.xlabel('Categoría')
plt.ylabel('Valor')
plt.show()



```

```python
#HEATMAPS

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Crear una matriz de datos ficticios
np.random.seed(42)
data_matrix = np.random.rand(10, 10)

# Crear un heatmap de la matriz de correlación
sns.heatmap(data_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap de la Matriz de Correlación')
plt.show()

#BARPLOTS

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame ficticio
data = {'categoria': ['A', 'B', 'C', 'A', 'B', 'C'],
        'valor': [10, 15, 5, 8, 12, 7]}
df = pd.DataFrame(data)

# Crear un bar plot
sns.barplot(x='categoria', y='valor', data=df, palette='viridis')
plt.title('Bar Plot de Categorías y Valores')
plt.xlabel('Categoría')
plt.ylabel('Valor')
plt.show()


#count plots

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame ficticio
data = {'categoria': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'A', 'B', 'C']}
df = pd.DataFrame(data)

# Crear un count plot
sns.countplot(x='categoria', data=df, palette='pastel')
plt.title('Count Plot de Categorías')
plt.xlabel('Categoría')
plt.ylabel('Cantidad')
plt.show()

#BOX PLOTS

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame ficticio
data = {'categoria': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'A', 'B', 'C'],
        'valor': [10, 15, 5, 8, 12, 7, 20, 25, 18, 22]}
df = pd.DataFrame(data)

# Crear un box plot
sns.boxplot(x='categoria', y='valor', data=df, palette='pastel')
plt.title('Box Plot de Categorías y Valores')
plt.xlabel('Categoría')
plt.ylabel('Valor')
plt.show()


#FACET GRIDS

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame ficticio
data = {'categoria': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'A'],
        'valor1': [10, 15, 5, 8, 12, 7, 20, 25, 18, 22],
        'valor2': [25, 20, 30, 15, 10, 18, 22, 27, 13, 30]}
df = pd.DataFrame(data)

# Crear un Facet Grid
g = sns.FacetGrid(df, col='categoria', col_wrap=3, height=4)
g.map(sns.scatterplot, 'valor1', 'valor2', palette='viridis')
plt.suptitle('Facet Grid de Scatter Plots por Categoría')
plt.show()


```

**Series temporales**

```python
#Representacion series temporales

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Crear una serie temporal ficticia
fecha_inicio = pd.to_datetime('2022-01-01')
fecha_final = pd.to_datetime('2022-12-31')
indice_tiempo = pd.date_range(fecha_inicio, fecha_final)
datos = np.random.randn(len(indice_tiempo)).cumsum()
serie_temporal = pd.Series(datos, index=indice_tiempo)

# Utilizar Seaborn para representar la serie temporal
sns.lineplot(x=serie_temporal.index, y=serie_temporal.values)
plt.title('Visualización de Serie Temporal con Seaborn')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```

**Graficos 3D**

```python
#3d con seaborn

import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np

# Crear datos ficticios tridimensionales
np.random.seed(42)
datos_3d = pd.DataFrame({'X': np.random.rand(100),
                          'Y': np.random.rand(100),
                          'Z': np.random.rand(100)})

# Crear un gráfico de dispersión 3D con Seaborn y Matplotlib
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(datos_3d['X'], datos_3d['Y'], datos_3d['Z'], c='skyblue', s=50)
ax.set_title('Gráfico de Dispersión 3D con Seaborn y Matplotlib')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
